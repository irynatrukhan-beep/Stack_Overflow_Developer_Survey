## Завдання 1. Підрахунок загальної кількості респондентів

```python
df['ResponseId'].nunique()
print("Кількість респондентів:", df['ResponseId'].nunique())
```

## Завдання 2. Аналіз повноти відповідей респондентів

```python
# Отримуємо список питань з поля "qname"
questions = set(df1['qname'].unique())
```

```python
# Інтерсекція множин — спільні стовпці між схемою і датафреймом
common_cols = list(questions & set(df.columns))
```

```python
# Фільтруємо df лише по цих стовпцях
df_filtered = df[common_cols]

# Підраховуємо респондентів без жодного пропуску
full_responses = df_filtered.dropna().shape[0]

print("Кількість респондентів, які відповіли на всі питання:", full_responses)
```

## Завдання 3. Статистичний аналіз досвіду респондентів

```python
mean_exp = df['WorkExp'].mean()
median_exp = df['WorkExp'].median()
mode_exp = df['WorkExp'].mode()[0]

print(f"Середнє значення досвіду роботи:  {mean_exp:.1f} років")
print(f"Медіана досвіду роботи:           {median_exp:.1f} років")
print(f"Мода досвіду роботи:              {mode_exp:.1f} років")
```

## Завдання 4. Аналіз віддаленої роботи

```python
# Переглядаємо унікальні значення поля RemoteWork
print(df['RemoteWork'].unique())
```

```python
# Фільтруємо лише тих хто працює повністю віддалено
remote_count = df[df['RemoteWork'] == 'Remote'].shape[0]

print(f"Кількість респондентів, які працюють віддалено: {remote_count}")
```

## Завдання 5. Визначення популярності Python

```python
df['worked_with_python'] = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
df['worked_with_python']
```

```python
# sum() рахує кількість True (True = 1, False = 0)
python_count = df['worked_with_python'].sum()
python_count
```

```python
# Загальна кількість респондентів
total = df['ResponseId'].nunique()
# Розраховуємо відсоток
python_pct = (python_count / total) * 100

print(f"Кількість респондентів, які використовують Python: {python_count}")
print(f"Відсоток респондентів, які програмують на Python: {python_pct:.1f}%")
```

## Завдання 7. Географічний аналіз компенсації Python-розробників

```python
# Фільтруємо лише Python-розробників
df_python = df[df['worked_with_python'] == True]

# Видаляємо рядки де відсутня зарплата
df_python = df_python.dropna(subset=['ConvertedCompYearly'])
```

```python
# Групуємо по країнах і рахуємо середню та медіанну компенсацію
py_progers_by_country = (
    df_python.groupby('Country')['ConvertedCompYearly']
    .agg(
        mean_progers='mean',
        median_progers='median'
    )
    .reset_index()
    .sort_values('mean_progers', ascending=False)
    .round(2)
)

py_progers_by_country
```

## Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів

```python
top5_education = (
    df.dropna(subset=['ConvertedCompYearly'])
    .sort_values('ConvertedCompYearly', ascending=False)
    .head(5)
    [['ResponseId', 'EdLevel', 'ConvertedCompYearly']]
    .reset_index(drop=True)  # оновлюємо індекси з 0 до 4
)

top5_education
```

## Завдання 9. Аналіз популярності Python по віковим категоріям 

```python
# Групуємо по віковій категорії
python_by_age = (
    df.groupby('Age')['worked_with_python']
    .mean()
    .reset_index()
    .rename(columns={'worked_with_python': 'python_pct'})
    .sort_values('python_pct', ascending=False)
    .reset_index(drop=True)
)
# Перетворюємо частку на відсотки і округлюємо
python_by_age['python_pct'] = (python_by_age['python_pct'] * 100).round(1)

python_by_age
```
